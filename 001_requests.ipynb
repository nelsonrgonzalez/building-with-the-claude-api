```python
from anthropic import Anthropic

client = Anthropic()
model = "claude-sonnet-4-0"
```

```python
message = client.messages.create(
    model=model,
    max_tokens=1000,
    messages=[
        {
            "role": "user",
            "content": "What is quantum computing? Answer in one sentence"
        }
    ]
)
message = client.messages.create(
    model=model,
    max_tokens=1000,
    messages=[
        {
            "role": "user",
            "content": "Write another sentence"
        }
    ]
)
message.content[0].text
```

```python
def add_user_message(messages, text):
    user_message = {"role": "user", "content": text}
    messages.append(user_message)

def add_assistant_message(messages, text):
    assistant_message = {"role": "assistant", "content": text}
    messages.append(assistant_message)


def chat(messages):
    message = client.messages.create(
        model=model,
        max_tokens=1000,
        messages=messages
    )
    return message.content[0].text
```

```python
# make a starting list of messages
messages = []

# add in the initial user question of "What is quantum computing? Answer in one sentence"
add_user_message(messages, "What is quantum computing? Answer in one sentence")

# passs the list of messages into 'chat' to get an answer
answer = chat(messages)

# take the answer and add it as an assistant message into our list
add_assistant_message(messages, answer)

# add in the user's follow-up question
add_user_message(messages, "Write another sentence")

# call chat again with teh list of messages to get a final answer
answer = chat(messages)

answer
```
